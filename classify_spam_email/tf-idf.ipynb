```python
from gensim.parsing.preprocessing import strip_non_alphanum, strip_multiple_whitespaces,preprocess_string, split_alphanum, strip_short, strip_numeric
```

```python
res = {}
words = []
with open('spam.txt', 'r', encoding = 'utf-8') as f:
    for line in f:
        line = strip_non_alphanum(line).lower().strip()
        token = line.split(" ")
        for tok in token:
            if tok in res.keys():
                res[tok] = res[tok] + 1
            else:
                res.update({tok : 1})
        word = {}
        for tok in token:
            if tok in word.keys():
                word[tok] = word[tok] + 1
            else:
                word.update({tok : 1})
        words.append(word)
# print(res)
print(len(res))
print(words)
```

```python
def computeTF(wordDict, words):
    tfDict = {}
    wordsCount = len(words)
    for word, count in wordDict.items():
        tfDict[word] = count/float(wordsCount)
    return tfDict
```

```python
tdfdoc = []
for word in words[0 : 10]:
    tdfdoc.append(computeTF(res, word))
print(tdfdoc)
```
